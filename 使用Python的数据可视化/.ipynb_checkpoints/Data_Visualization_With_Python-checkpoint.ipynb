## 简介

本文使用[WHO的新冠病毒统计数据](https://covid19.who.int/table)，日期范围为一个国家开始有第一个New_cases的时间到2020年8月。

文中将使用python的matplotlib做数据可视化分析，根据可能出现的不同分析场景需求画出不同的可视化图。

在业务中的实际应用主要对应数据分析中的**描述分析**这一步骤。

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.dates
from datetime import datetime
```

## 1. 数据预处理
数据导入与简单的数据清洗，将WHO提供下载的Excel中的Date_reported改成标准的datetime，YYYY-MM-DD格式。

```python
df = pd.read_csv('WHO-COVID-19-global-data.csv')
df.head()
```

```python
df['Date_reported'] = pd.to_datetime(df.Date_reported, dayfirst = True)
df.head()
```

## 2. 数据可视化

### 2.1 简单绘图

此处假设我们希望对美国新冠病毒从2月到8月的发展趋势有一个宏观的了解，那么我们可以直接以时间为横坐标，累计病例为纵坐标作图。

```python
US_DATA = df.loc[df['Country'] == 'United States of America']
US_DATA.head()
```

```python
plt.plot(US_DATA['Date_reported'], US_DATA['Cumulative_cases'])
plt.title('Cumulative COVID-19 Cases in United States')
plt.xlabel('Date Reported')
plt.ylabel('Cumulative Cases')

plt.autoscale()
plt.show()
```

此时我们可能还希望将美国的疫情发展趋势与一个其他国家对比。

比如英国，那么我们就可以画第二条线在同一个的图像上。

```python
UK_DATA = df.loc[df['Country'] == 'The United Kingdom']
UK_DATA.head()
```

```python
plt.plot(US_DATA['Date_reported'], US_DATA['Cumulative_cases'], label = 'US')
plt.plot(UK_DATA['Date_reported'], UK_DATA['Cumulative_cases'], label = 'UK')

plt.title('Cumulative COVID-19 Cases in United States vs United Kingdom')
plt.xlabel('Date Reported')
plt.ylabel('Cumulative Cases')

plt.legend()
plt.autoscale()
plt.show()
```

### 2.2 对数图

在一些特定的场景中，我们可能希望知道数据在何时达到了某一数量级（如十万、百万），此时可以用对数图plt.semilogy解决。

如下图，我们可以很直观地看到美国新冠感染人数在什么时候达到了千级、万级。

```python
plt.semilogy(US_DATA['Date_reported'], US_DATA['Cumulative_cases'])

plt.title('Cumulative COVID-19 Cases in United States')
plt.xlabel('Date Reported')
plt.ylabel('Cumulative Cases')

plt.autoscale()
plt.show()
```

### 2.3 泡式图

泡式图的优势在于气泡的大小和颜色可以给图片提供更多维度的信息。

比如我们在看美国新冠感染数量的同时也关心美国的死亡数量以及死亡率（累计死亡 / 累计确诊），那么我们就可以通过气泡图来实现。

```python
mask = (df['Date_reported'].isin(['2020-02-01', '2020-03-01', '2020-04-01','2020-05-01','2020-06-01', '2020-07-01', '2020-08-01',
                                 '2020-02-15', '2020-03-15', '2020-04-15','2020-05-15','2020-06-15', '2020-07-15', '2020-08-15']))

US_Bubble = US_DATA.loc[mask]

death_rate = US_Bubble['Cumulative_deaths'] / US_Bubble['Cumulative_cases']

plt.scatter(US_Bubble['Date_reported'], US_Bubble['Cumulative_cases'], s = 20000*death_rate, alpha=0.5, c = US_Bubble['Cumulative_deaths'])

plt.title('COVID-19 Cases & Deaths in United States')

plt.xlabel('Date Reported')
plt.ylabel('Cumulative Cases')

plt.xlim(['2020-01-01', '2020-09-01'])
plt.show()
```

上图以半个月为一个统计周期，时间为x轴，感染人数为y轴，死亡人数为气泡颜色，死亡率为气泡大小

我们不难看出，虽然死亡人数持持续增加趋势，但是死亡率的最高点其实在5-6月份，之后有所回落。

### 2.4 直方图

在了解某一数据分布的的时候我们会希望用到直方图来更好地了解数据的分布形态。（如研究数据是否符合正态分布）

下例中假设我们想知道新冠病毒在数据获取日2020.8.17日的全球死亡率的分布情况。

```python
Latest = df.loc[df['Date_reported'] == '2020-08-17']
Latest['Death_Rate'] = Latest['Cumulative_deaths'] / Latest['Cumulative_cases']
Latest.head()
```

```python
plt.hist(Latest['Death_Rate'], edgecolor='k', alpha=0.35, histtype='bar', rwidth=0.8)

plt.title('Global COVID-19 Deaths Rates Distribution')

plt.xlabel('Death Rate')
plt.ylabel('Number of Countries')

plt.show()
```

### 2.5 散点图

上面同样的问题也可以通过散点图来得到答案，同时，散点图也可以在检测聚类算法（如K-MEANS）的效果中使用。

```python
plt.scatter(Latest['Country_code'], Latest['Death_Rate'], alpha=0.6, c='red')

plt.title('Global COVID-19 Deaths Rates Distribution Scatter Plot')

plt.ylabel('Death Rate')
plt.xticks([])
plt.show()
```

### 2.6 条形图

在解决有些业务问题时，我们可能会需要进行不同 区域 / 用户年龄层 / 产品类别 之间的互相对比，此时条形图就能很好地反映出某一个类别对于我们关心的指标的影响。

下例中，我们假设现在想知道6月的第一周，全世界不同地区（按照世界卫生组织的分类一共7个）的新冠新增病例情况。

对于WHO_region分类不清晰的可以参考[这个链接](https://en.wikipedia.org/wiki/List_of_WHO_regions)

```python
mask = (df['Date_reported'] >= '2020-06-01') & (df['Date_reported'] <= '2020-06-07')

JUNE1STWEEK = df.loc[mask]

Region_DATA = JUNE1STWEEK.loc[:,['Date_reported','WHO_region', 'New_cases']].groupby(['Date_reported', 'WHO_region'], as_index=False)['New_cases'].sum()
Region_DATA.head(10)
```

```python
AFRO, AMRO, EMRO, EURO, WPRO, SEARO, Other =  [Region_DATA.loc[Region_DATA['WHO_region'] == 'AFRO'],
Region_DATA.loc[Region_DATA['WHO_region'] == 'AMRO'],
Region_DATA.loc[Region_DATA['WHO_region'] == 'EMRO'],
Region_DATA.loc[Region_DATA['WHO_region'] == 'EURO'],
Region_DATA.loc[Region_DATA['WHO_region'] == 'WPRO'],
Region_DATA.loc[Region_DATA['WHO_region'] == 'SEARO'],
Region_DATA.loc[Region_DATA['WHO_region'] == 'Other']]
```

```python
labels = Region_DATA['Date_reported'].dt.date.unique()

x = np.arange(len(labels))
width = 0.12

fig, ax = plt.subplots()
rects1 = ax.bar(x - 3 * width, AFRO['New_cases'], width, label='AFRO')
rects2 = ax.bar(x - 2 * width, AMRO['New_cases'], width, label='AMRO')
rects3 = ax.bar(x - width, EMRO['New_cases'], width, label='EMRO')
rects4 = ax.bar(x, EURO['New_cases'], width, label='EURO')
rects5 = ax.bar(x + width, WPRO['New_cases'], width, label='AMRO')
rects6 = ax.bar(x + 2 * width, SEARO['New_cases'], width, label='EMRO')
rects7 = ax.bar(x + 3 * width, Other['New_cases'], width, label='Other')

ax.set_ylabel('New_cases')
ax.set_title('Regional Total New Cases at the First Week of June')
ax.set_xticklabels(labels)
ax.legend()
fig.set_size_inches(25, 6)


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)
autolabel(rects5)
autolabel(rects6)
autolabel(rects7)

plt.show()

```

### 2.7 饼图

下例中我们想了解7大地区新冠病例日新增数量占全球新增数量的比例，此时饼图就能很好地对数据的比例有一个呈现。

```python
August = df.loc[df['Date_reported'] == '2020-08-01']
August_Regional = August.loc[:,['Date_reported','WHO_region', 'New_cases']].groupby(['Date_reported', 'WHO_region'], as_index=False)['New_cases'].sum()

labels = August_Regional['WHO_region'].unique()
NewCases = August_Regional['New_cases']

labels
```

```python
colors = ['red','yellowgreen','lightskyblue','yellow','purple','pink','peachpuff','orange']
explode = (0, 0.1, 0, 0, 0, 0, 0)
fig1, ax1 = plt.subplots()

ax1.pie(NewCases, labels=labels, colors=colors,
        explode = explode, autopct='%1.1f%%',
        shadow=True, startangle=90, pctdistance = 0.8)

fig1.set_size_inches(8, 8)
plt.axis('equal')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

colors = ['red','yellowgreen','lightskyblue','yellow','purple','pink','peachpuff','orange']
explode = (0, 0.1, 0, 0, 0, 0, 0)

March = df.loc[df['Date_reported'] == '2020-03-01']
March_Regional = March.loc[:,['Date_reported','WHO_region', 'New_cases']].groupby(['Date_reported', 'WHO_region'], as_index=False)['New_cases'].sum()

labels = March_Regional['WHO_region'].unique()
Mar_NewCases = March_Regional['New_cases']

fig = plt.figure() 

gs = gridspec.GridSpec(1, 2)


ax1 = plt.subplot(gs[0, 0])


ax1.pie(Mar_NewCases, labels=labels, colors=colors,
        explode = explode, autopct='%1.1f%%',
        shadow=True, startangle=90, pctdistance = 0.8)


April = df.loc[df['Date_reported'] == '2020-04-01']
April_Regional = April.loc[:,['Date_reported','WHO_region', 'New_cases']].groupby(['Date_reported', 'WHO_region'], as_index=False)['New_cases'].sum()

labels = April_Regional['WHO_region'].unique()
April_NewCases = April_Regional['New_cases']

ax2 = plt.subplot(gs[0, 1])


ax2.pie(April_NewCases, labels=labels, colors=colors,
        explode = explode, autopct='%1.1f%%',
        shadow=True, startangle=90, pctdistance = 0.8)

ax1.set_title('Global New Cases Distribution at the beginning of March')
ax2.set_title('Global New Cases Distribution at the beginning of April')

fig.add_subplot(ax1)
ax1.legend()
fig.add_subplot(ax2)
ax2.legend()
fig.set_size_inches(20, 10)
```

在需要的时候，我们也可以将两个月的饼图放在一起做对比。

### 2.8 箱线图

箱线图是利用数据中的五个统计量：最小值、第一四分位数、中位数、第三四分位数与最大值来描述数据的一种方法。

下例中我们通过箱线图分析2020年8月1日，AFRO, AMRO, WPRO, SEARO中各国的新冠总计死亡率的分布情况。

```python
August1st = df.loc[df['Date_reported'] == '2020-08-01']
August1st['Death_Rate'] = August1st['Cumulative_deaths'] / August1st['Cumulative_cases']

AFRO, AMRO, EMRO, EURO, WPRO, SEARO, Other =  [August1st.loc[August1st['WHO_region'] == 'AFRO'],
August1st.loc[August1st['WHO_region'] == 'AMRO'],
August1st.loc[August1st['WHO_region'] == 'EMRO'],
August1st.loc[August1st['WHO_region'] == 'EURO'],
August1st.loc[August1st['WHO_region'] == 'WPRO'],
August1st.loc[August1st['WHO_region'] == 'SEARO'],
August1st.loc[August1st['WHO_region'] == 'Other']]
```

```python
data = [AFRO['Death_Rate'], AMRO['Death_Rate'], WPRO['Death_Rate'], SEARO['Death_Rate']]


plt.boxplot(data)
plt.show()
```

```python
import seaborn as sns

sns.set()
plt.figure(figsize=(10,8))
sns.boxplot(x = 'WHO_region', y = 'Death_Rate', data = August1st)
plt.show()
```

如果此处不需要指定比较哪几个Region的话，Seaborn有自动group的功能，也很好用

### 2.10 热力图

```python
mask = (df['Date_reported'].isin(['2020-02-01', '2020-03-01', '2020-04-01','2020-05-01','2020-06-01', '2020-07-01', '2020-08-01']))

Month_Start = df.loc[mask].groupby(['Date_reported', 'WHO_region'], as_index=False)['New_cases'].sum()
Month_Start['Date_reported'] = Month_Start['Date_reported'].dt.date


Month_Start.head(10)
```

```python
plt.figure(figsize=(15,10))

result = Month_Start.pivot(index='WHO_region', columns='Date_reported', values='New_cases')
sns.heatmap(result, annot=True, fmt="g",)
```

## 3. 添加标注

一个好的数据可视化报告一定不会只有数据本身，还要对数据有一定业务意义上的标注，这样数据分析报告的阅读者才能直观地理解报告的主要内容。

```python
US_DATA = df.loc[df['Country'] == 'United States of America']

plt.rcParams["figure.figsize"] = (15,8)

plt.plot(US_DATA['Date_reported'], US_DATA['Cumulative_cases'], label = 'U.S.', c = 'red')
plt.title('Cumulative COVID-19 Cases in United States and Canada')
plt.xlabel('Date Reported')
plt.ylabel('Cumulative Cases')

CA_DATA = df.loc[df['Country'] == 'Canada']
plt.plot(CA_DATA['Date_reported'], CA_DATA['Cumulative_cases'], label = 'Canada', c = 'yellow')


plt.annotate('First Circuit Break \n 2020.3.9', xy=('2020-03-09', 10000), xytext=('2020-02-20', 1500000),
            arrowprops=dict(facecolor='blue', shrink=2, width = 3), bbox = dict(boxstyle="round", fc="0.8"))

plt.annotate('Second Circuit Break \n 2020.3.16', xy=('2020-03-16', 10000), xytext=('2020-04-10', 1200000),
            arrowprops=dict(facecolor='blue', shrink=2, width = 3), bbox = dict(boxstyle="round", fc="0.8"))

plt.legend()
plt.show()


```

上图的例子中对被可视化的数据【美国 & 加拿大 2月到8月累计新冠病例数量统计】做了如下备注：

1. 用Seaborn的Default格式加了网格。
2. 用matplotlib的annotation标注了3.9，3.16这两个美股发生熔断的日期。
3. 添加标题、x轴、y轴名称等信息以及标注了红色代表美国，黄色代表加拿大。
