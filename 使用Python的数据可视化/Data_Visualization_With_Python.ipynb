```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.dates
from datetime import datetime
```

```python
df = pd.read_csv('WHO-COVID-19-global-data.csv')
df.head()
```

```python
df['Date_reported'] = pd.to_datetime(df.Date_reported, dayfirst = True)
df.head()
```

```python
US_DATA = df.loc[df['Country'] == 'United States of America']
US_DATA.head()
```

```python
plt.plot(US_DATA['Date_reported'], US_DATA['Cumulative_cases'])
plt.title('Cumulative COVID-19 Cases in United States')
plt.xlabel('Date Reported')
plt.ylabel('Cumulative Cases')

plt.autoscale()
plt.show()
```
